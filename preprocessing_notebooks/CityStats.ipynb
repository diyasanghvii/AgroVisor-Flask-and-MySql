```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(r'/content/drive/MyDrive/531-Datasets/CityStats/MedianHouseholdIncome2015.csv', encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
df1.head()
```

```python
df2 = pd.read_csv(r'/content/drive/MyDrive/531-Datasets/CityStats/PercentOver25CompletedHighSchool.csv', encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
df2.head()
```

```python
df3 = pd.read_csv(r'/content/drive/MyDrive/531-Datasets/CityStats/PercentagePeopleBelowPovertyLevel.csv', encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
df3.head()
```

```python
df4 = pd.read_csv(r'/content/drive/MyDrive/531-Datasets/CityStats/ShareRaceByCity.csv', encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
df4.head()
```

```python
# Rename "Geographic area" to "Geographic Area" in df4
df4 = df4.rename(columns={"Geographic area": "Geographic Area"})
```

```python
# Perform the first merge
merged_df = pd.merge(df1, df2, on=['City','Geographic Area'], how='inner')

# Perform the second merge
merged_df = pd.merge(merged_df, df3, on=['City','Geographic Area'], how='inner')

# Perform the third merge
merged_df = pd.merge(merged_df, df4, on=['City','Geographic Area'], how='inner')

merged_df.head()
```

```python
# Rename columns in merged_df
merged_df = merged_df.rename(columns={
    'Geographic Area': 'State',
    'City': 'City',
    'Median Income': '%medianIncome',
    'percent_completed_hs': '%highSchoolAbove25',
    'poverty_rate': '%belowPoverty',
    'share_white': '%White',
    'share_black': '%Black',
    'share_native_american': '%NativeAmerican',
    'share_asian': '%Asian',
    'share_hispanic': '%Hispanic'
})
```

```python
# Get row count before dropping
row_count_before = merged_df.shape[0]

# Replace '(X)' with NaN
merged_df.replace('(X)', np.nan, inplace=True)

# Replace '-' with NaN
merged_df.replace('-', np.nan, inplace=True)

# Drop rows with any NaN value
merged_df.dropna(inplace=True)

# Get row count after dropping
row_count_after = merged_df.shape[0]

# Display row count before and after
print(f"Row count before dropping: {row_count_before}")
print(f"Row count after dropping: {row_count_after}")
```

```python
merged_df['ID'] = ['CS' + str(i) for i in range(1, len(merged_df) + 1)]
```

```python
# Save the DataFrame to a CSV file
merged_df.to_csv('CityStats.csv', index=False)
```
